```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\mamat\OneDrive\Documents\Udemy\House_price_prediction\HousePrice.csv")
df
```

```python
x=df[["Aera"]]
x
```

```python
y=df[["Price"]]
y
```

```python
plt.title("aera vs price")
plt.xlabel("aera")
plt.ylabel("price")
plt.scatter(x,y,color="green",s=60,ec="purple")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
pred =model.predict(x)
pred
```

```python
model.predict([[2600]])
```

```python
model.predict([[455444]])
```

```python
plt.title("aera vs price")
plt.xlabel("aera")
plt.ylabel("price")
plt.scatter(x,y,color="green",s=60,ec="purple")
plt.plot(x,pred)
plt.show()
```
